```python
import autograd.numpy as np
from autograd import grad
import matplotlib.pyplot as plt
# import numpy as np
```

```python
# for single variable
def my_function(x):
  return np.sin(x)


# find the gradient of my_function
gradient_func = grad(my_function)
gradient = gradient_func(2.0)

print(gradient)
```

```python
# for two variable

def myfunc(X):
  x,y=X
  return x**2+y**2


# find the gradient of my_function
gradient_func = grad(myfunc)
gradient = gradient_func([3.0,2.0])

print(gradient)
print(type(gradient))
```

```python
import math
x_lst = []
for i in range(250):
  x_lst.append(i/10)

```

```python
# symbolic differentation of f(x) = sin(x)
y_sym = []
for i in range(len(x_lst)):
  a = math.cos(x_lst[i])
  y_sym.append(a)
  
```

```python
# numerical differentation of graph f(x) = sin(x)
dx = 0.5
y_num = []
for i in range(len(x_lst)):
  b = (math.sin(x_lst[i]+dx)-math.sin(x_lst[i]))/dx
  y_num.append(b)
```

```python
# Differentation using autograd :
y_auto=[]
for i in range(len(x_lst)):
  gradient_func = grad(my_function)
  gradient = gradient_func(x_lst[i])
  y_auto.append(gradient)
  
```
