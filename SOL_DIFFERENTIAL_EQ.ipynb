COMPLETE TUTORIAL ON ANALYTICAL AND EULER METHOD:-

```python
import plotly.express as px
import math
import plotly.graph_objects as go
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Analytical solution for x(t) = e**t

lst_t = []
for i in range(51):
  lst_t.append(i/10)
lst_x = []  
for i in lst_t:
  lst_x.append(math.exp(i))

# Command for plotting graph
#fig = px.line(x=lst_t,y=lst_x)
#fig.show()
plt.plot(lst_t,lst_x)


```

```python
# Numerical solution(Eulers method)
lst_t_e = []
lst_x_e= []
x_hat = 1
t = 0.0
while t < 5:
  lst_t_e.append(t)
  x_hat = x_hat + x_hat*(0.1)
  lst_x_e.append(x_hat)
  t += 0.1

plt.plot(lst_t,lst_x)

```

```python
# Comparision b/w analytical and numerical solution to analyse and see the error

plt.plot(lst_t, lst_x, label= 'Analytical solution')
plt.legend()
plt.plot(lst_t_e, lst_x_e, label = 'Solution By Euler')
plt.legend()
```

```python
# Error Measurement  between analytical and numerical method
# error = mod of (analitical solution - numerical solytion)/analitical solution
dt = []
for i in range(51):
  dt.append(i/1000)

# For finding Error
error = []
for j in range(51):
  err = (lst_x[j] - lst_x_e[j])/lst_x[j]
  error.append(err)

plt.plot(dt, error)
plt.show()
```

```python
# Modified Euler method (midpoint method) x1 = x0 + dt(x0 + dt*x0/2)
number_mid = [2]
dt = 0.1
for j in range(50):
  x1 = number_mid[j] + dt*(number_mid[j] + dt*number_mid[j]/2)
  number_mid.append(x1)
plt.plot(lst_t, number_mid)
plt.show

```

```python
# measure error between Euler(midpoint method) and analytical method
dt = []
for i in range(51):
  dt.append(i/10)

# for finding error
error = []
for j in range(51):
  err = abs(lst_x[j] - number_mid[j])/number_mid[j]
  error.append(err)

plt.plot(dt, error)
plt.show()
```
