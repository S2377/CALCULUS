This tutorial for implementation of various integration methods and plottings of their errors(error vs partition width) as partition width goes very very small. Integration methods are  Sympson's integration and trapezoidal. These all are numerical methods used to approximate definite integral of a function.

#Sympson's integration:
It is a numerical method used to approximate the definite integral of a function.This method is based on approximating the area under the curve of a function by dividing the area into a series of smaller sections, and approximating the area of each section using a quadratic function.Simpson's integration method is particularly effective when the function being integrated is relatively smooth and has few sharp changes in direction.The formula for Simpson's integration:
$$\int_{a}^{b}x dx = \frac{b - a}{6}( f(a) + f(b))$$
Where $f(x)$ is the function being integrated, $a$ and $b$ are the lower and upper limits of integration.

# Trapezoidal method of Integration
This is a numerical integration technique that approximates the definite integral of a function by approximating the area under the curve using a trapezoidal shape. Formula is given below:

```python
# Importing required modules and pacakages.
from scipy.integrate import quad
import matplotlib.pyplot as py
```

```python
# Defining the function given for integration.
def f(x):
    return x ** 2 + 2 * x + 1
limit = [float(input(' Lower limit: ')), float(input(' Upper limit: '))]      # lower limits and upper limit.
```

```python
# Calculating integration using python built-in module.
result, error = quad(f, limit[0], limit[1])
```

```python
# Implementation of all types of methods of integration.
sym_Int = []
trap_Int = []
del_XList = []                     
int_val = limit[1] - limit[0]
partition = 10

while partition < 10000:
    deltaX = int_val / (partition)
    del_XList.append(deltaX)
    tagPart = limit[0] + deltaX / 2         # Tagging of partitions.
    sympsonSum = 0
    trapezoidalSum = 0
    for i in range(partition):
        lowerLimit = limit[0] + i * deltaX
        u = lowerLimit + deltaX
        m = lowerLimit + u
        trapezoidalSum = trapezoidalSum + (f(lowerLimit) + f(u)) * deltaX / 2
        sympsonSum = sympsonSum + (f(lowerLimit) / 3 + f(m / 2) * (4 / 3) + f(u) / 3) * deltaX / 2
        tagPart += deltaX
    sym_Int.append(abs(sympsonSum - result))
    trap_Int.append(trapezoidalSum - result)
    partition *= 2
```

```python
# Command for drawing the graph
py.show()
py.plot(del_XList, sym_Int)
py.show()
py.plot(del_XList, trap_Int)
```
