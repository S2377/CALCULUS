RIEMANN INTEGRAL

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 + 2*x + 1

def riemann_sum(f, a, b, n, method='left'):
    """
    Calculates the Riemann integral sum of a function f over the interval [a, b]
    using a specified method: 'left', 'right', or 'midpoint'.
    """
    dx = (b - a) / n
    x = np.linspace(a, b, n+1)
    if method == 'left':
        x = x[:-1]
    elif method == 'right':
        x = x[1:]
    elif method == 'midpoint':
        x = x[:-1] + dx / 2
    else:
        raise ValueError("Invalid method specified. Choose 'left', 'right', or 'midpoint'.")
    return np.sum(f(x) * dx)


```

```python
# Define the integration interval [a, b]
a = 0
b = 5
```

```python
# Define the number of subintervals for the Riemann sum
n_values = np.arange(1, 101)  # Vary n from 1 to 100

# Calculate the Riemann sum and error for each value of n
results = []
errors = []
for n in n_values:
    result = riemann_sum(f, a, b, n, method='left')
    error = abs(result - (f(b) - f(a)))
    results.append(result)
    errors.append(error)
```

```python

# Plot the error graph
plt.plot(n_values, errors)
plt.xlabel('Number of subintervals (n)')
plt.ylabel('Error')
plt.title('Riemann Sum Error vs. Number of Subintervals')
plt.show()

# Calculate the definite integral using quad() function from scipy
from scipy.integrate import quad

result, error = quad(f, a, b)

# Print the result and error
print("The result of the integration is: ", result)
print("The error estimate is:", error)
```
