# **The Perimetre Of An Ellipse**

The ellipse is given by the parametric equitions,

$$x = acos(t)$$  $$y = bsin(t) $$where t is parametric angle and  a is the length of the major axis, and b is the length of the minor axis.

Now the differential arc lenght for the ellipse is given by

$$\frac{dr}{dt}=\sqrt{(\frac{dx}{dt})^{2}+(\frac{dy}{dt})^{2}}dt$$

Therefore the perimeter of the ellipse is given by integral

$$Perimeter=\int_{0}^{2\pi}\sqrt{a^{2}sin^{2}(t)+b^{2}cos^{2}(t)}dt$$

```python
# required module and pacakages
from scipy.integrate import quad
import numpy as np
import math
```

```python
# define the function to integrate(numerical integration)
def f(x):
  return x**2 + 2*x +1

# integrate the fun from 0 to 2
result,error = quad(f,0,2)
print('The result of the integration is ',result)
print('The error estimate is',error)
```

```python
a_lst = [i/10 for i in range(20,40) for j in range(10,20)]
x = [i/10 for i in range(10,20)]
b_lst = x + x + x + x + x + x + x + x + x + x + x + x + x + x + x + x + x + x + x + x
# print(a_lst)
# print(len(a_lst))
# print(b_lst)
# print(len(b_lst))
```

```python
f_lst = []
for i,j in zip(a_lst,b_lst):
  def f(t):
    return (i**2*math.sin(t)**2 + j**2*math.cos(t)**2)**0.5

  # integrate the fun from 0 to 2
  result,error = quad(f,0,2*math.pi)
  f_lst.append(result)
  #print(i,j,result)
# print('The result of the integration is ',result)
# print('The error estimate is',error)
print(f_lst)
```

```python
# command for drawing the graph
import plotly.express as px
fig = px.scatter_3d(x=a_lst, y=b_lst, z=f_lst)
fig.show()
```

# comparison with few approximate formula
some formula to approximate the perimeter of ellipse


1. Approximation based on the arithmetic mean:
$$P ≈ 2π(a + b)/2$$

2. Approximation based on the geometric mean:

$$P ≈ 2π√(ab)/2$$


3. Ramanujan approximation:
$$P ≈2\pi\sqrt{(\frac{a^{2}+b^{2}}{2})}$$

4. Ramanujan's formula for better approximation :
$$P ≈ π[(a + b) - √((3a+b)(a+3b)]$$


```python

def f(t):
    return (a**2*(math.sin(t))**2 + b**2*(math.cos(t))**2)**0.5

# Initializing empty lists
b_list = []
f_list = []
error_am = []
error_gm = []
error_ramanujan = []
x_list = []
```

```python
# Setting values of a and b and iterating over range of b
a = 4
b = 4
for i in range(100):
    # Appending value of b to b_list
    b_list.append(b)
    # # Integrating the function f from 0 to 2*pi using quad function
    result = quad(f,0,2*math.pi)
    f_list.append(result[0])

    # # Calculating perimeters using different approximate formula formulas
    perimeter_am = 2*math.pi*((a+b)/2)
    perimeter_gm = 2*math.pi*((a*b)**0.5)
    perimeter_ramanujan = 2*math.pi*((a**2 + b**2)/2)**0.5
    
    # perimeter3 =  math.pi*(6*a/5 + 3*b/4)

    # # Calculating errors for each perimeter approximation and appending to error lists
    error_am.append((abs(result[0] - perimeter_am)/result[0])*100)
    error_gm.append((abs(result[0] - perimeter_gm)/result[0])*100)
    error_ramanujan.append((abs(result[0] - perimeter_ramanujan)/result[0])*100)
    #  Appending ratio of b to a to x_list
    x_list.append(b/a)
    # Increasing value of b by 0.1 for next iteration
    b += 0.1
```

```python
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatter(name = 'arithmetic mean', x =x_list, y = error_am))
fig.add_trace(go.Scatter(name = 'geometric mean', x =x_list, y = error_gm))
fig.add_trace(go.Scatter(name = 'Ramujan_approximation', x =x_list, y = error_ramanujan))
fig.show()
```
